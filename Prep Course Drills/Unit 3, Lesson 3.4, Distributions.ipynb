```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
my_logistic = np.random.logistic(3, 2, 100)
plt.hist(my_logistic, color='m')
plt.title("Figure 1: Logistic Distribution")
plt.axvline(my_logistic.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(my_logistic.mean() + my_logistic.std(), color='b', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(my_logistic.mean()-my_logistic.std(), color='b', linestyle='dashed', linewidth=2) 

plt.show()
```

```python
my_exponential = np.random.exponential(2, 100)
plt.hist(my_exponential, color='m')
plt.title("Figure 2: Exponential Distribution")
plt.axvline(my_exponential.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(my_exponential.mean() + my_exponential.std(), color='b', linestyle='dashed', linewidth=2)

#Add another vertical line at two sds above the mean.
plt.axvline(my_exponential.mean() + 2*my_exponential.std(), color='y', linestyle='dashed', linewidth=2)


# Add a vertical line at one standard deviation below the mean.
plt.axvline(my_exponential.mean()-my_exponential.std(), color='b', linestyle='dashed', linewidth=2) 

plt.show()
```

```python
my_lognormal = np.random.lognormal(1, 1, 100)
plt.hist(my_lognormal, color='m')
plt.title("Figure 3: Log-normal Distribution")
plt.axvline(my_lognormal.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(my_lognormal.mean() + my_lognormal.std(), color='b', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(my_lognormal.mean()-my_lognormal.std(), color='b', linestyle='dashed', linewidth=2) 

plt.show()
```

```python
my_chisquare = np.random.chisquare(2,100)
plt.hist(my_chisquare, color='m')
plt.title("Figure 4: Chi-square Distribution")
plt.axvline(my_chisquare.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at one and two standard deviations above the mean.
plt.axvline(my_chisquare.mean() + 2*my_chisquare.std(), color='y', linestyle='dashed', linewidth=2)

plt.axvline(my_chisquare.mean() + my_chisquare.std(), color='b', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(my_chisquare.mean() -my_chisquare.std(), color='b', linestyle='dashed', linewidth=2) 

plt.show()
```

```python
my_wald = np.random.wald(10,5, 100)
plt.hist(my_wald, color='m')
plt.title("Figure 5: Wald Distribution")
plt.axvline(my_wald.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(my_wald.mean() + my_wald.std(), color='b', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(my_wald.mean() -my_wald.std(), color='b', linestyle='dashed', linewidth=2) 

plt.show()
```

```python
my_poisson = np.random.poisson(5, 100)
plt.hist(my_poisson, color='m')
plt.title("Figure 6: Poisson Distribution")
plt.axvline(my_poisson.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(my_poisson.mean() + my_poisson.std(), color='b', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(my_poisson.mean() -my_poisson.std(), color='b', linestyle='dashed', linewidth=2) 

plt.show()
```

Each of the distributions above is non-normal, and includes some skew either to the right or the left. The logistic and poisson distributions above can be described usefully with the mean and the standard deviation, since the data does have a central tendancy, and most of the observations are within +/- one standard deviation of the mean, but they are asymmetrical, unlike the normal distribution, skewing to the right. 

The descriptive statistics (mean and standard deviation) of the other distributions, which are asymettrical with very long right tails, are not very useful, since the vast majority of the data is centered just to the left of the mean. This shape is particularly striking in the Wald and log-normal distributions, where most of the data is concentrated in the first bin, with a steep droppoff to the right. The exponential and chi-square distributions show a gentler slope after the first bin, and a slight rise around +2 standard deviations from the mean.  

```python
normal1 = np.random.normal(5,.5, 100)
normal2 = np.random.normal(10,1, 100)
normal3 = normal1 + normal2

plt.hist(normal3, color='m')
plt.title("Figure 7: Sum of Two Random Normal Distributions")
plt.axvline(normal3.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(normal3.mean() + normal3.std(), color='b', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(normal3.mean() -normal3.std(), color='b', linestyle='dashed', linewidth=2) 

plt.show()
```

In Figure 7, the sum of two normal distributions is depicted in a histogram. The result is a non-normal, bimodal distribution. It is more symmetrical than, for example, a Wald distribution, because the left and right tails include the tails of the original two normal distributions. The mean and standard deviation are useful in this case: the concentration of data within +/- one standard deviation of the mean includes the peaks around the means of the two normal distributions that make up the distribution. The peaks at the means of the original two distributions represent real factors at play in the summed distribution. 

In a real analysis, the two peaks could be very informative. For instance (to take an example from my own teaching experience), if the grades of a population of students are bimodally distributed, it could indicate that the class includes two subpopulations of students with different learning needs.
