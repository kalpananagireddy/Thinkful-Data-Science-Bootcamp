```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## Comparing two random binomial distributions

```python
#Create two random binomial distributions. 
pop1 = np.random.binomial(10, 0.2, 10000)
pop2 = np.random.binomial(10,0.5, 10000) 

#Plot the two distributions.
plt.hist(pop1, alpha=0.5, label='Population 1') 
plt.hist(pop2, alpha=0.5, label='Population 2') 
plt.legend(loc='upper right') 
plt.axvline(pop1.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(pop1.mean() + pop1.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(pop1.mean()-pop1.std(), color='b', linestyle='dashed', linewidth=2) 
plt.axvline(pop2.mean(), color='darkorange', linestyle='solid', linewidth=2)
plt.axvline(pop2.mean() + pop2.std(), color='darkorange', linestyle='dashed', linewidth=2)
plt.axvline(pop2.mean()-pop2.std(), color='darkorange', linestyle='dashed', linewidth=2) 
plt.title("Histrograms of two random binominal distributions")
plt.show()

print('The mean of population 1 is {}.'.format(pop1.mean()))
print('The sd of population 1 is {}.'.format(pop1.std()))
print('The mean of population 2 is {}.'.format(pop2.mean()))
print('The sd of population 2 is {}.'.format(pop2.std()))
```

### Comparing samples (n=100)

```python
#Create two samples (n=100) from each population.
sample1 = np.random.choice(sample1, 100, replace=True)
sample2 = np.random.choice(sample2, 100, replace=True)

#Plot the two samples.
plt.hist(sample1, alpha=0.5, label='Sample 1') 
plt.hist(sample2, alpha=0.5, label='Sample 2') 
plt.legend(loc='upper right') 
plt.axvline(sample1.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(sample1.mean() + sample1.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(sample1.mean()-sample1.std(), color='b', linestyle='dashed', linewidth=2) 
plt.axvline(sample2.mean(), color='darkorange', linestyle='solid', linewidth=2)
plt.axvline(sample2.mean() + sample2.std(), color='darkorange', linestyle='dashed', linewidth=2)
plt.axvline(sample2.mean()-sample2.std(), color='darkorange', linestyle='dashed', linewidth=2) 
plt.title("Histrograms of samples (n=100) from two random binominal distributions")
plt.show()

#Calculate the means and standard deviations of the samples.

print('The mean of sample 1 is {}.'.format(sample1.mean()))
print('The sd of sample 1 is {}.'.format(sample1.std()))
print('The mean of sample 2 is {}.'.format(sample2.mean()))
print('The sd of sample 2 is {}.'.format(sample2.std()))
```

The sample means are very close to the population means; the standard deviations appear to be a bit larger than those of the populations, because the effect of variance is greater when the size of the sample is smaller. 

### Effect of sample size on estimates of population mean and sd

If the sample size is increased to 1000, we expect, given the Central Limit Theorem, that mean and standard deviations will approach those of the population. On the other hand, if the sample size is decreased to 20, there is more variance in the sample, so its mean and standard deviation may diverge more from those of the population. 

```python
sample1a = np.random.choice(pop1, 1000, replace=True)
sample2a = np.random.choice(pop2, 1000, replace=True)

plt.hist(sample1a, alpha=0.5, label='Sample 1 (n=1000)') 
plt.hist(sample2a, alpha=0.5, label='Sample 2 (n=1000)') 
plt.legend(loc='upper right') 
plt.axvline(sample1a.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(sample1a.mean() + sample1a.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(sample1a.mean()-sample1a.std(), color='b', linestyle='dashed', linewidth=2) 
plt.axvline(sample2a.mean(), color='darkorange', linestyle='solid', linewidth=2)
plt.axvline(sample2a.mean() + sample2a.std(), color='darkorange', linestyle='dashed', linewidth=2)
plt.axvline(sample2a.mean()-sample2a.std(), color='darkorange', linestyle='dashed', linewidth=2) 
plt.title("Histrograms of samples (n=1000) from two random binominal distributions")
plt.show()

print('The mean of sample 1 (n=1000) is {}.'.format(sample1a.mean()))
print('The sd of sample 1 (n=1000) is {}.'.format(sample1a.std()))
print('The mean of sample 2 (n=1000) is {}.'.format(sample2a.mean()))
print('The sd of sample 2 (n=1000) is {}.'.format(sample2a.std()))
```

```python
sample1b = np.random.choice(pop1, 20, replace=True)
sample2b = np.random.choice(pop2, 20, replace=True)

plt.hist(sample1b, alpha=0.5, label='Sample 1 (n=20)') 
plt.hist(sample2b, alpha=0.5, label='Sample 1 (n=20)') 
plt.legend(loc='upper right') 
plt.axvline(sample1b.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(sample1b.mean() + sample1a.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(sample1b.mean()-sample1a.std(), color='b', linestyle='dashed', linewidth=2) 
plt.axvline(sample2b.mean(), color='darkorange', linestyle='solid', linewidth=2)
plt.axvline(sample2b.mean() + sample2b.std(), color='darkorange', linestyle='dashed', linewidth=2)
plt.axvline(sample2b.mean()-sample2b.std(), color='darkorange', linestyle='dashed', linewidth=2) 
plt.title("Histrograms of samples (n=20) from two random binominal distributions")
plt.show()

print('The mean of sample 1 (n=20) is {}.'.format(sample1b.mean()))
print('The sd of sample 1 (n=20) is {}.'.format(sample1b.std()))
print('The mean of sample 2 (n=20) is {}.'.format(sample2b.mean()))
print('The sd of sample 2 (n=20) is {}.'.format(sample2b.std()))
```

As predicted, the sample means and standard deviations are noticeably closer to the population means and standard deviations when the sample sizes are larger. 

## Effect of population p on comparison of means 

```python
#Adjust p for population 1, take new samples, and inspect the mean and standard deviations.
pop1 = np.random.binomial(10, 0.3, 10000)
pop2 = np.random.binomial(10,0.5, 10000)

plt.hist(pop1, alpha=0.5, label='Population 1') 
plt.hist(pop2, alpha=0.5, label='Population 2') 
plt.legend(loc='upper right') 
plt.axvline(pop1.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(pop1.mean() + pop1.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(pop1.mean()-pop1.std(), color='b', linestyle='dashed', linewidth=2) 
plt.axvline(pop2.mean(), color='darkorange', linestyle='solid', linewidth=2)
plt.axvline(pop2.mean() + pop2.std(), color='darkorange', linestyle='dashed', linewidth=2)
plt.axvline(pop2.mean()-pop2.std(), color='darkorange', linestyle='dashed', linewidth=2) 
plt.title("Histrograms of two random binominal distributions")
plt.show()


print('The mean of population 1 is {}.'.format(pop1.mean()))
print('The sd of population 1 is {}.'.format(pop1.std()))
print('The mean of population 2 is {}.'.format(pop2.mean()))
print('The sd of population 2 is {}.'.format(pop2.std()))

sample1 = np.random.choice(pop1, 100, replace=True)
sample2 = np.random.choice(pop2, 100, replace=True)

print('The mean of sample 1 is {}.'.format(sample1.mean()))
print('The sd of sample 1 is {}.'.format(sample1.std()))
print('The mean of sample 2 is {}.'.format(sample2.mean()))
print('The sd of sample 2 is {}.'.format(sample2.std()))
```

The t-statistic and p-value for the difference between sample 1 and sample 2 are reported below.

```python
#Compute the t-statistic and p-value for the difference between the samples.
from scipy.stats import ttest_ind
print(ttest_ind(sample2, sample1, equal_var=False))
```

## Comparing samples from a logseries distribution

```python
pop1 = np.random.logseries(0.2, 10000)
pop2 = np.random.logseries(0.5, 10000) 

# Let’s make histograms for the two groups. 

plt.hist(pop1, alpha=0.5, label='Population 1') 
plt.hist(pop2, alpha=0.5, label='Population 2') 
plt.legend(loc='upper right') 
plt.axvline(pop1.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(pop1.mean() + pop1.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(pop1.mean()-pop1.std(), color='b', linestyle='dashed', linewidth=2) 
plt.axvline(pop2.mean(), color='darkorange', linestyle='solid', linewidth=2)
plt.axvline(pop2.mean() + pop2.std(), color='darkorange', linestyle='dashed', linewidth=2)
plt.axvline(pop2.mean()-pop2.std(), color='darkorange', linestyle='dashed', linewidth=2) 
plt.title("Histrograms of two random logseries")
plt.show()

print('The mean of population 1 is {}.'.format(pop1.mean()))
print('The sd of population 1 is {}.'.format(pop1.std()))
print('The mean of population 2 is {}.'.format(pop2.mean()))
print('The sd of population 2 is {}.'.format(pop2.std()))

sample1 = np.random.choice(pop1, 100, replace=True)
sample2 = np.random.choice(pop2, 100, replace=True)

print('The mean of sample 1 is {}.'.format(sample1.mean()))
print('The sd of sample 1 is {}.'.format(sample1.std()))
print('The mean of sample 2 is {}.'.format(sample2.mean()))
print('The sd of sample 2 is {}.'.format(sample2.std()))
from scipy.stats import ttest_ind
print(ttest_ind(sample2, sample1, equal_var=False))
```

In the logseries distributions depicted in the histogram above, the sample means and standard deviations accurately represent the population values. Although the distribution has a different shape from the binominal distributions tested earlier in this drill, the extreme clustering of the data around the means still shows a difference between the two populations. 

The t-statistic and p-value detect this difference with a sample size of 100, even though we are examining a different type of distribution from the binomial examples above. According to the Central Limit Theorem, we would expect the t-statistic to increase and the p-value to decrease as the size of the sample approached the size of the population; on the other hand, the t-statistic should decrease and the p-value should increase if the sample size shrinks.
